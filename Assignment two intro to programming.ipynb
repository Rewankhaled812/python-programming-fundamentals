1.	Write function pow() that calculates the power, which takes base and exponent as arguments.

```python
def pow(base,exponent):
    power=1
    for i in range (exponent):
        power*=base
    return power
b=int(input("enter the base: "))
e=int(input("enter the exponent: "))
print(pow(b,e))

```

7.	Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

```python
num=int(input("Enter the number: "))
def prime(num):
    if num <= 1:
        return False
    
    for i in range(2, int(num**0.5) +1):
        if num % i == 0:
            return False
    
    return True
```

8.	Write a Python class Restaurant with attributes like menu_items, book_table, and customer_orders, and methods like add_item_to_menu, book_tables, and customer_order. (Note: Use dictionaries and lists to store the data.)
Perform the following tasks now:
-	Now add items to the menu.
-	Make table reservations.
-	Take customer orders.
-	Print the menu.
-	Print table reservations.
-	Print customer orders.


```python
class Restaurant:
    def __init__(self):
        self.menu_items=[]
        self.book_tables={}
        self.customer_orders={}
    def add_item_to_menu(self,menu_item):
        if menu_item not in self.menu_items:
            self.menu_items.append(menu_item)

        
    def book_table(self,table_num: int,name:str)->None:
        self.book_tables[table_num]=name

    def customer_order(self,name,items:list)->None: #type hinting
        self.customer_orders[name]=items
    
    def prtint_menu(self):
        return f"menu items: {self.menu_items}"
    def print_table_reservations(self):
        return f"Table reservations {self.book_tables}"
    def print_customer_orders(self):
        return f"Customer orders: {self.customer_orders}"
    
```

```python
r1=Restaurant()
r1.add_item_to_menu("pizza")
r1.add_item_to_menu("ice coffe")
r1.book_table(1,"Leen")
r1.customer_order("Aya",["pizza", "ice coffee"])
print(r1.prtint_menu())
print(r1.print_table_reservations())
print(r1.print_customer_orders())
```
