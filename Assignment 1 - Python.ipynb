## Problem 7: Palindrome and Armstrong Number Check

Write a program to accept number from the user and check if the given number is palindrome or Armstrong or not.

**Examples:**
- Palindrome: 12321 (reads the same forwards and backwards)
- Armstrong: 153 (1³ + 5³ + 3³ = 1 + 125 + 27 = 153)

````python
number=int(input("Enter the number : "))
original_num=number
rev_num=0
sum_power=0
````

````python
def palindrom(number,original_num,rev_num):
    while number!=0:
        last=number%10
        rev_num=rev_num*10+last
        number=number//10
    if original_num==rev_num:
        return "palindrome"
    else:
        return "NOT palindrome"
````

````python
def armstrong(number,original_num,sum_power):
    num_of_digits=len(str(number))
    while number!=0:
        last=number%10
        sum_power=sum_power+(last**num_of_digits)
        number=number//10
    if original_num == sum_power:
        return "Armstrong"
    else:
        return "NOT Armstrong"
````

````python
palind_result=palindrom(number, original_num, rev_num)
armstromg_result=armstrong(number,original_num,sum_power)
print(f"The number is {palind_result} and {armstromg_result}.")
````

## Problem 9: Series Sum Calculation

Write a program to find the sum of the following series: 1 + 4 – 9 +16 – 25 + 36 ...........n terms

Note: The input is n (number of terms)

**Example:**
- Input: 5 → Output: 1 + 4 – 9 +16 – 25 = -13

````python
n = int(input("Enter number of terms: "))

total = 1
for i in range(2, n + 1):
    term = i * i
    if i % 2 == 0:
        total += term
    else:
        total -= term

print("Sum of the series is:", total)
````

## Problem 11: Input Type Processing

Write a program that ask user to enter input (text or number or number with text):
- In case the input is number get the summation of its digits
  - Example: 123 → 6
- In case the input is text remove the spaces
  - Example: "the car" → "thecar"
- In case number with text split them, number and text each individually
  - Example: "23 st" → [23] [st]

````python
user_input = input("Enter input (text, number, or number with text): ")

if user_input.isdigit():
    # Input is a number
    digit_sum = 0
    for i in user_input:
        digit_sum += int(i)
    
    print(f"Summation of digits: {digit_sum}")

elif user_input.replace(' ', '').isalpha():
    # Input is text
    text_without_spaces = user_input.replace(" ", "")
    print(f"Text without spaces: {text_without_spaces}")

else:
    # Input is a mix of numbers and text
    number_part = ''
    text_part = ''
    for char in user_input:
        if char.isdigit():
            number_part += char
        elif char.isalpha():
            text_part += char

    print(f"Number part: {number_part}")
    print(f"Text part: {text_part}")
````

## Problem 12: Average of Tuple of Tuples

Write a Python program to calculate the average value of the numbers in a given tuple of tuples.

**Original Tuple:**
```
((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 32), (1, 2, 3, 4))
```

**Expected Output:**
```
Average value of the numbers of the said tuple of tuples: [30.5, 34.25, 27.0, 23.25]
```

````python
tub=((10, 10, 10, 12),
      (30, 45, 56, 45), 
      (81, 80, 39, 32), 
      (1, 2, 3, 4))
````

````python
def avgOfTuple(tub):
    for i in zip(*tub):
        summation= sum(i)
        avr=summation/len(i)
        print(avr)
````

## Problem 15: Extract Elements with High Frequency

Write a Python program to extract elements with Frequency greater than element K.

**Example:**
- Input: test_list = [4, 6, 4, 3, 3, 4, 3, 4, 6, 6], K = 2
- Output: [4, 3, 6]

````python
# Example input
test_list = [4, 6, 4, 3, 3, 4, 3, 4, 6, 6]
K = 2
````

````python
# Count the frequency of each element
frequency = {}
for element in test_list:
    if element in frequency:
        frequency[element] += 1
    else:
        frequency[element] = 1
````

````python
# Extract elements with frequency greater than element K's frequency
result = []

if K in frequency.keys():
    limit = frequency[K]
else:
    limit = 0

for num in frequency:
    count = frequency[num]
    if count > limit:
        result.append(num)

print(result)
````
